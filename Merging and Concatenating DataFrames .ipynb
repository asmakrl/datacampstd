# Concatenating DataFrames

```python
# dataframes having the same columns
df1 = pd.DataFrame({'Name': ['Asma', 'Sarah', 'Chakib', 'Ilyes'],
                    'Age': [29, 28, 21, 18],
                    'Gender': ['F', 'F', 'M', 'M']}
                  )

df2 = pd.DataFrame({'Name': ['Mohamed', 'Islam', 'Meryem'],
                    'Age': [31, 22, 19],
                    'Gender': ['M', 'M', 'F']}
                  )
```

```python
df1
```

```python
df2
```

```python
df = pd.concat([df1, df2])
df= df.reset_index()
df = df.drop(['index'], axis = 1)
df
```

```python
df = df1.append(df2)
df=df.reset_index()
df = df.drop(['index'], axis = 1)
df
```
